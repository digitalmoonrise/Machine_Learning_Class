#building a sentiment classifier

```python
products['rating'].show
```

```python
products['rating'].show(view='Categorical')
```

##Define what's a positive and a negative

```python
#ignore all 3* reviews
products = products[products['rating']!=3]
```

```python
#positive sentiment = 4* or 5* rating
products['sentiment'] = products['rating'] >= 4
```

```python
products.head()
```

#Training our sentiment classifier

```python
train, test = products.random_split(.8, seed = 0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train, 
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set = test)
```

#evaluate the sentiment model

```python
sentiment_model.evaluate(test, metric = 'roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```

```python
products['rating'].show(view='Catagorical')
```

#applying the learned model to understand sentiment for the Giraffe

```python
giraffe_reviews['predicted_sentiment'] = sentiment_model.predict(giraffe_reviews, output_type='probability')
```
